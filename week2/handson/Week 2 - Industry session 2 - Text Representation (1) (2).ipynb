<font color=darkblue><h2>**Let's recollect some core concepts of Text Representation**</h2></font>

<h3>Text representation is a core concept in Natural Language Processing (NLP) that involves transforming text data into a numerical format that can be processed and analyzed by machine learning algorithms.
</h3> ===================================================================================================================

#### There are different techniques for representing text data, including:

* **One-hot encoding**: One-hot encoding is a common technique used in Natural Language Processing (NLP) to represent categorical data, such as words or tokens, as binary vectors. In the context of NLP, one-hot encoding is typically applied to create vector representations of words in a corpus.

* **Bag-of-Words (BoW)**: This technique represents text as a set of individual words, disregarding grammar and word order. Each word is assigned a unique identifier, and the frequency of each word in a document is recorded. This results in a sparse matrix where each row represents a document, and each column represents a word.

* **TF-IDF (Term Frequency-Inverse Document Frequency)**: This technique is similar to BoW, but instead of counting the frequency of each word, it weighs each word based on how important it is to a document. Words that appear frequently in a document but infrequently in other documents are considered more important.

* **Word Embeddings**: This technique represents each word as a high-dimensional vector, where each dimension corresponds to a different semantic feature. Word embeddings are learned through neural network models, such as Word2Vec and GloVe, which are trained on large text corpora.
>> * **Word2Vec**: It was introduced by researchers at Google and has become widely used in natural language processing (NLP) tasks.

>>>> * Word2Vec offers two main algorithms to learn word embeddings: Continuous Bag-of-Words (CBOW) and Skip-gram. Both algorithms are trained on large text corpora and aim to capture the semantic and syntactic relationships between words.

>> * **GloVe**: GloVe (Global Vectors for Word Representation) is an unsupervised learning algorithm used for obtaining vector representations of words. It was developed by researchers at Stanford University. The main idea behind GloVe is to capture the meaning of words by analyzing their co-occurrence statistics in a large corpus of text.

>> * **FastText**: FastText is another popular word embedding technique that extends the concept of word representations beyond individual words to include subword information. It was developed by Facebook AI Research (FAIR) and has gained significant popularity in the NLP community.

Overall, text representation is a critical component of NLP, as it enables machine learning models to process and analyze textual data effectively.

```python
import warnings
warnings.filterwarnings("ignore")

from sklearn.feature_extraction.text import CountVectorizer

# Sample documents
documents = ["This is the first document.",
             "This is the second document.",
             "And this is the third one.",
             "Is this the first document?"]

# Create a CountVectorizer object
vectorizer = CountVectorizer()

# Fit and transform the documents
X = vectorizer.fit_transform(documents)

# Print the resulting BoW representation
print(X.toarray())
```


<font color=darkblue><h3 align='center'> Bag of N-grams using scikit-learn library :</h3></font>

image.png

* The bag of n-grams model is useful for capturing important combinations of words or characters in a document, regardless of their position. It can be used as input for various machine learning algorithms, such as text classification, sentiment analysis, and information retrieval, to extract meaningful patterns and features from text data.

```python
# Bag of 2-gram (bigram)-ngram_range=(2, 2)

from sklearn.feature_extraction.text import CountVectorizer

# Sample corpus
corpus = [
    "This is the first document.",
    "This document is the second document.",
    "And this is the third one.",
    "Is this the first document?"
]

# Initialize the CountVectorizer with n-gram range
ngram_vectorizer = CountVectorizer(ngram_range=(2, 2))

# Fit the vectorizer on the corpus and transform the corpus into a bag of n-grams matrix
bag_of_ngrams = ngram_vectorizer.fit_transform(corpus)

# Get the feature names (n-grams)
feature_names = ngram_vectorizer.get_feature_names()

# Print the bag of n-grams matrix
print(bag_of_ngrams.toarray())

# Print the feature names
print(feature_names)
```

```python
# ngram_range=(1, 2). This means that it will consider both unigrams (individual words)
# and bigrams (two-word combinations) as features. You can adjust the ngram_range parameter to include different n-gram combinations.

from sklearn.feature_extraction.text import CountVectorizer

# Sample corpus
corpus = [
    "This is the first document.",
    "This document is the second document.",
    "And this is the third one.",
    "Is this the first document?"
]

# Initialize the CountVectorizer with n-gram range
ngram_vectorizer = CountVectorizer(ngram_range=(1, 2))

# Fit the vectorizer on the corpus and transform the corpus into a bag of n-grams matrix
bag_of_ngrams = ngram_vectorizer.fit_transform(corpus)

# Get the feature names (n-grams)
feature_names = ngram_vectorizer.get_feature_names()

# Print the bag of n-grams matrix
print(bag_of_ngrams.toarray())

# Print the feature names
print(feature_names)
```

- The resulting bag of n-grams matrix now includes both unigrams and bigrams.

 <font color=darkblue><h3 align='center'>2. TF-IDF using scikit-learn:</h3></font>


![image.png](https://miro.medium.com/v2/resize:fit:943/1*Xw19xjYH_oyaIoX2znlPjw.png)

* TF is calculated using the following formula:
  TF(term, document) = (Number of occurrences of the term in the document) / (Total number of terms in the document)

* IDF is calculated using the following formula:
  IDF(term, collection) = log((Total number of documents in the collection) / (Number of documents containing the term))

* TF-IDF is calculated using the following formula:
  TF-IDF(term, document, collection) = TF(term, document) * IDF(term, collection)

```python
from sklearn.feature_extraction.text import TfidfVectorizer

# Sample documents
documents = ["This is the first document.",
             "This is the second document.",
             "And this is the third one.",
             "Is this the first document?"]

# Create a TfidfVectorizer object
vectorizer = TfidfVectorizer()

# Fit and transform the documents
X = vectorizer.fit_transform(documents)

# Print the resulting TF-IDF representation
print(X.toarray())
```

 #### To perform word embedding in NLP using Continuous Bag-of-Words (CBOW) and Skip-gram models, you can use the gensim library in Python. Gensim provides an efficient implementation of these algorithms for word2vec

```python
from gensim.models import Word2Vec

# Sample corpus
corpus = [
    ['I', 'love', 'NLP'],
    ['NLP', 'is', 'awesome'],
    ['Python', 'is', 'fun']
]

# Train Word2Vec model
model = Word2Vec(corpus, size=100, window=5, min_count=1, workers=4)

# Get word vector for a specific word
word = 'NLP'
word_vector = model.wv[word]

# Print word vector
print("{}Word Vector:{}".format('\033[1m', '\033[0m'),'\n', word_vector)

```

```python
from gensim.models import Word2Vec

# Sample corpus
corpus = [
    ['I', 'love', 'NLP'],
    ['NLP', 'is', 'awesome'],
    ['Python', 'is', 'fun']
]

# Train CBOW model
cbow_model = Word2Vec(corpus, sg=0, size=100, window=2, min_count=1, workers=4)  #sg=0 indicates CBOW.

# Train Skip-gram model
skipgram_model = Word2Vec(corpus, sg=1, size=100, window=2, min_count=1, workers=4)  #sg=1 indicates Skip-gram.

# Get word vectors
word = 'NLP'
cbow_word_vector = cbow_model.wv[word]
skipgram_word_vector = skipgram_model.wv[word]

# Print word vectors
print( "{}CBOW Word Vector:{}".format('\033[1m', '\033[0m'),'\n',cbow_word_vector)
print ("--------------------------------------------------------------------")
print("{}Skip-gram Word Vector:{}".format('\033[1m', '\033[0m'),'\n', skipgram_word_vector)
```

 <font color=darkblue><h3 align='center'>4. One-hot encoding using the scikit-learn library</h3></font>

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.feature_extraction.text import CountVectorizer

# Sample corpus
corpus = [
    'I love NLP',
    'NLP is awesome',
    'I enjoy working with NLP'
]

# Step 1: Tokenization and vocabulary creation
vectorizer = CountVectorizer()
X = vectorizer.fit_transform(corpus)
vocabulary = vectorizer.get_feature_names_out()

# Step 2: Index assignment
word_to_index = {word: index for index, word in enumerate(vocabulary)}

# Step 3: One-hot encoding
encoder = OneHotEncoder(categories='auto')
one_hot_encoded = encoder.fit_transform([[word_to_index[word]] for word in vocabulary])

# Print the vocabulary and one-hot encoded vectors
print("{}Vocabulary:{}".format('\033[1m', '\033[0m'), vocabulary)
print("{}One-hot encoded vectors:{}".format('\033[1m', '\033[0m'))
for i, word in enumerate(vocabulary):
    print(word, one_hot_encoded[i].toarray()[0])
```

 <font color=darkblue><h3 align='center'>5. GloVe Word Embedding </h3></font>

**To use GloVe word embeddings in Python, you can follow these steps:**

1. Download GloVe pre-trained word vectors:

* GloVe provides pre-trained word vectors in different dimensions and trained on different corpus sizes. You can download the pre-trained vectors from the GloVe website: https://nlp.stanford.edu/projects/glove/.
* Choose the appropriate vector size and download the corresponding file. For example, "glove.6B.100d.txt" contains 100-dimensional word vectors trained on a 6 billion token corpus.

2. Load the GloVe word vectors into memory:

* Once downloaded, you need to load the word vectors into your Python script. You can load them into a dictionary where the keys are the words, and the values are the corresponding vectors.

```python
import numpy as np

# Load GloVe word vectors into a dictionary. Please give your glove path adress here.
glove_path = r"C:\Users\Vedavathi\Downloads\glove.6B\glove.6B.100d.txt"

embedding_dim = 100
word_vectors = {}
with open(glove_path, encoding="utf-8") as file:
    for line in file:
        values = line.split()
        word = values[0]
        vector = np.asarray(values[1:], dtype="float32")
        word_vectors[word] = vector

# Example usage: get word vector for a specific word
word = "study"
vector = word_vectors.get(word)
print(f"Word vector for '{word}':'\n' {vector}")
```

- The word vectors are stored in a dictionary called word_vectors.

* You can then use the word_vectors dictionary to retrieve word vectors for specific words. In the example, we retrieve the word vector for the word "study" and print it.

**Note: Make sure to adjust the file path (glove_path) and embedding dimension (embedding_dim) according to the GloVe file you downloaded.**

**Preprocessing**

Text preprocessing steps include a few essential tasks to further clean the available text data. It includes tasks like:-

**1. Stop-Word Removal** : In English, words like a, an, the, as, in, on, etc. are considered as stop-words so according to our requirements we can remove them to reduce vocabulary size as these words don't have some specific meaning

**2. Lower Casing** : Convert all words into the lower case because the upper or lower case may not make a difference for the problem.
And we are reducing vocabulary size by doing so.

**3. Stemming** : Stemming refers to the process of removing suffixes and reducing a word to some base form such that all different variants of that word can be represented by the same form (e.g., study and studying are both reduced to study).

**4. Tokenization** : NLP software typically analyzes text by breaking it up into words (tokens) and sentences.

Pre-processing of the text is not the main objective of this notebook that's why I am just covering a few basic steps in a brief

```python
# Stemming
ps = PorterStemmer()
# Initializing Lists
corpus = []
words = []
for i in range(0, len(data)):
    # Removing characters other than letters
    review = re.sub('[^a-zA-Z]', ' ', str(data["clean_text"][i]))
    # Lowering the case all the text
    review = review.lower()
    # Splitting into words
    review = review.split()
    # Applying Stemming
    stemmed = [ps.stem(word) for word in review if not word in stopwords.words('english')]
    # Joining words
    review = ' '.join(stemmed)
    # Appending all tweets to a list after preprocessing
    corpus.append(review)
    # Appending all words for word embeddings
    words.append(stemmed)
```
